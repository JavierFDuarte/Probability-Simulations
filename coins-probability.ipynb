```python
#Supongamos que se arrojan n coins. ¿Cuál es la probabilidad de obtener x caras?
#Caracter probabilístico dado por la distibución binomial
#Simulación numérica
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Documentación matplotlib: https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.hist.html
#En particular lo de rwidth es lo que me importó, es el ancho relativo de las columnas respecto al de los bins
```

```python
n_coins=7
n_muestras=5000
results = [ ]
bins = [ ]
for j in range(1,n_muestras+1):
    suma = 0.
    for i in range(1,n_coins+1):
        x = np.random.random_integers(0,1)
        suma = suma+float(x)
    results.append(suma)

#print(results)

for i in range(0,n_coins+2):
    bins.append(float(i)-0.5)
#print(bins)
grafico=plt.hist(results,bins=bins,rwidth=0.5, edgecolor = "black",density=True) 
plt.show(grafico)
k=0
for i in grafico[0]:
    print("P(",k,")=",i)
    k=k+1
```
